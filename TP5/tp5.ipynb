```python
def get_frames(capture, n_samples):
    start_frame = capture.get(cv.CAP_PROP_POS_FRAMES)
    frames = []

    for i in range(n_samples):
        capture.set(cv.CAP_PROP_POS_FRAMES, start_frame + i)
        frames.append(capture.read()[1])

    capture.set(cv.CAP_PROP_POS_FRAMES, start_frame)
    return np.array(frames)

def get_frames_rand(capture, n_samples, start, end):
    start_frame = capture.get(cv.CAP_PROP_POS_FRAMES)
    random_indices = [random.randint(start, end) for _ in range(n_samples)]
    frames = []

    for index in random_indices:
        capture.set(cv.CAP_PROP_POS_FRAMES, index)
        frames.append(capture.read()[1])

    capture.set(cv.CAP_PROP_POS_FRAMES, start_frame)
    return np.array(frames)


def write_filename(frame, filename):
    cv.rectangle(frame, (10, 2), (300, 20), (255, 255, 255), -1)
    cv.putText(
        frame,
        filename,
        (15, 15),
        cv.FONT_HERSHEY_SIMPLEX,
        0.5,
        (0, 0, 0),
    )
    return frame
```

```python
def BgSubs_cv2(video_path, output_path, method="MOG2"):
    if method == "MOG2":
        bg_sub = cv.createBackgroundSubtractorMOG2()
    elif method == "KNN":
        bg_sub = cv.createBackgroundSubtractorKNN()
    else:
        return

    start_time = time.time()
    cap = cv.VideoCapture(video_path)
    w = int(cap.get(cv.CAP_PROP_FRAME_WIDTH))
    h = int(cap.get(cv.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv.CAP_PROP_FPS))
    video_writer = cv.VideoWriter(output_path, 0, fps, (w, h))

    while True:
        _, frame = cap.read()

        if frame is None:
            break

        mask = bg_sub.apply(frame)
        result = write_filename(mask, os.path.basename(output_path))
        video_writer.write(cv.cvtColor(result, cv.COLOR_GRAY2BGR))
    elapsed = time.time() - start_time

    print(f"Tiempo empleado: {elapsed:.3f} segundos")
    cv.destroyAllWindows()
    cap.release()
    video_writer.release()
```
